## Filtering the Course name column for Analytics Courses

```python
filt=['data analytics','business analytics','data analysis','business intelligence']
```

```python
da=df[
    (df['Course Name'].str.contains('|'.join(filt),case=False)) &\
    (df['Subject Name'].isin(['Business and Economics','Computer Science']))
  ]
da.head(5)
```

```python
da.shape
```

## There are 17 courses offered in the field of analytics

```python
da['Required German language'].value_counts()
```

### None of the courses in data analytics require german language

## Checking for free courses

```python
df_free=da[~da['Tuition Fees'].str.isdigit()]
```

```python
len(df_free)
```

## There are only 8 courses in analytics that have zero tuition fees

```python
df_free
```

```python
df_free=df_free.reset_index(drop=True)
df_free
```

### Dropping the unneccesary columns

```python
df_free.drop(columns=['Required German language','Required English language'],inplace=True)
```
